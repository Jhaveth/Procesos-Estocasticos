## Ejercicio 2.1.1
Considere el ejemplo 2.4 anterior, a partir de este realice una simulación de la trayectoria del proceso para los primeros 100 días, posteriormente construya un codigo en python de  una gráfica de la proporción de días lluviosos (para inicia las simulaciones considere como estados iniciales que los dos días previos se presentó lluvia)




#Solucion:
Comenzamos por Definir la matriz de transición:
Dado que hemos definido cuatro estados posibles (0, 1, 2 y 3) para representar diferentes combinaciones de días lluviosos y no lluviosos en los dos días previos, necesitamos construir una matriz de transición de probabilidades. Cada entrada de la matriz representará la probabilidad de transición de un estado a otro. Dado el enunciado, podemos definir la matriz de transición de la siguiente manera:

```python
import numpy as np

# Definir la matriz de transición
transicion_matri = np.array([
    [0.7, 0.3, 0.0, 0.0],  # Probabilidades desde el estado 0
    [0.5, 0.0, 0.5, 0.0],  # Probabilidades desde el estado 1
    [0.4, 0.0, 0.0, 0.6],  # Probabilidades desde el estado 2
    [0.2, 0.0, 0.0, 0.8]   # Probabilidades desde el estado 3
])
```

Luego, para simular la trayectoria del proceso para los primeros 100 días:
Para simular la trayectoria del proceso durante 100 días, comenzaremos desde un estado inicial en el que los dos días previos llovió (estado 0) y luego usaremos la matriz de transición para determinar los estados futuros en cada paso.

```python
# Estado inicial: los dos días previos llovió (estado 0) (numero semilla)
estado_inicial = 0

# Simulación de la trayectoria del proceso durante 100 días
num_dias = 100
dias_lluviosos = [estado_inicial]

for _ in range(num_dias - 1):
    sig_estado = np.random.choice([0, 1, 2, 3], p=transicion_matri[dias_lluviosos[-1]])
    dias_lluviosos.append(sig_estado)
```

Por ultimo necesitamos construir el código para graficar la proporción de días lluviosos y crear una grafica:


```python
import matplotlib.pyplot as plt

# Calcular la proporción de días lluviosos en cada paso de tiempo
prop_dias_lluv = [state in [0, 1] for state in dias_lluviosos]

# Crear una lista de días (1, 2, ..., 100) para el eje X
days = list(range(1, num_dias + 1))

# Crear la gráfica de la proporción de días lluviosos
plt.plot(days, prop_dias_lluv, marker='o', linestyle='-')
plt.xlabel('Días')
plt.ylabel('Proporción de Días Lluviosos')
plt.title('Proporción de Días Lluviosos en los Primeros 100 Días')
plt.grid(True)
plt.show()
```
